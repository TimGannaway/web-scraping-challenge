```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.news_db
collection = db.articles
```

```python
# URL of page to be scraped
url = 'https://globalvoices.org/'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Retrieve the parent divs for all articles
results = soup.find_all('div', class_='post-excerpt-container')

# loop over results to get article data
for result in results:
    # scrape the article header 
    header = result.find('h3', class_='post-title').text.strip()
    
    # scrape the article news region
    region = result.find('span', class_='term-link term-link-category').text
    
    # scrape the date
    date = result.find('span', class_='datestamp')['title'] 
    
    # print article data
    print('-----------------')
    print(header)
    print(region)
    print(date)

    # Dictionary to be inserted into MongoDB
    post = {
        'header': header,
        'region': region,
        'date': date,
    }

    # Insert dictionary into MongoDB as a document
    collection.insert_one(post)
```

```python
# Display the MongoDB records created above
articles = db.articles.find()
for article in articles:
    print(article)
```
